# Kiểm tra tính nhất quán train với status

$|\pi_{segment\_id, date}(train) \cap \pi_{segment\_id, DATE(updated\_at)}(segment\_status)|= x$, 
$x\in \{|\pi_{segment\_id, date}(train)|, |\pi_{segment\_id, DATE(updated\_at)}(segment\_status)|\}$

$|\pi_{segment\_id, date}(train) \cap \pi_{segment\_id, FLOOR(updated\_at, 30m)}(segment\_status)|= y$, 
$x\in \{|\pi_{segment\_id, date}(train)|, |\pi_{segment\_id, FLOOR(updated\_at, 30m)}(segment\_status)|\}$

```python
status_df["time_bucket"] = status_df["updated_at"].dt.floor("30min")
```

```python
status_df["period"] = (
    "period_" +
    status_df["time_bucket"].dt.hour.astype(str) + "_" +
    status_df["time_bucket"].dt.strftime("%M")
)
```

```python
status_df["date"] = status_df["updated_at"].dt.strftime("%Y-%m-%d")
```

```python
train_date = set(train_df[["segment_id", "date"]].apply(tuple, axis=1))
status_date = set(status_df[["segment_id", "date"]].apply(tuple, axis=1))
```

```python
train_date = set(train_df[["segment_id", "date"]].apply(tuple, axis=1))
status_date = set(status_df[["segment_id", "date"]].apply(tuple, axis=1))

inter_train_status_date = train_date & status_date
print("Tập date trong train:", len(train_date))
print("Tập date trong status:", len(status_date))
print("Tập giao:", len(inter_train_status_date))
```

```python
train_period = set(train_df[["segment_id", "period"]].apply(tuple, axis=1))
status_period = set(status_df[["segment_id", "period"]].apply(tuple, axis=1))

inter_train_status_period = train_period & status_period
print("Tập date trong train:", len(train_period))
print("Tập date trong status:", len(status_period))
print("Tập giao:", len(inter_train_status_period))
```

```python
_train_status_period = train_period - status_period

for idx, v in enumerate(_train_status_period):
    if idx == 4:
        break
    print(v)
```
